```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Beispiel-Datensatz erstellen
data = {
    'ID': range(1, 21),
    'Alter': [35, 42, 28, 30, 45, 29, 34, 33, 50, 27, 36, 40, 32, 48, 25, 38, 43, 26, 41, 37],
    'Geschlecht': ['Männlich', 'Weiblich', 'Männlich', 'Weiblich', 'Männlich', 'Weiblich', 'Männlich', 
                   'Männlich', 'Weiblich', 'Männlich', 'Weiblich', 'Männlich', 'Weiblich', 'Männlich', 
                   'Weiblich', 'Männlich', 'Männlich', 'Weiblich', 'Männlich', 'Weiblich'],
    'Abteilung': ['Montage', 'Qualitätskontrolle', 'Lackierung', 'Montage', 'Endkontrolle', 'Montage', 
                  'Lackierung', 'Endkontrolle', 'Qualitätskontrolle', 'Montage', 'Lackierung', 
                  'Endkontrolle', 'Qualitätskontrolle', 'Montage', 'Endkontrolle', 'Lackierung', 
                  'Qualitätskontrolle', 'Montage', 'Endkontrolle', 'Lackierung'],
    'Produktionszeit': [5.5, 7.2, 6.8, 5.9, 8.0, 6.1, 5.7, 7.9, 7.0, 5.8, 6.5, 8.1, 7.4, 6.0, 5.6, 6.7, 7.3, 5.4, 7.5, 6.6],
    'Fehleranzahl': [1, 0, 2, 1, 0, 2, 1, 0, 1, 1, 3, 1, 0, 1, 0, 2, 0, 1, 1, 2]
}

df = pd.DataFrame(data)
```

### **Datentypen identifizieren:**
Numerische Daten: Alter, Produktionszeit, Fehleranzahl.
Kategoriale Daten: Geschlecht, Abteilung.


```python
df.style.hide(axis="index")
```

### **Grundlegende statistische Analyse:**
Berechne den Durchschnitt, Median, Minimum und Maximum für die numerischen Felder.

Beispiel: Durchschnittliche Produktionszeit, durchschnittliche Fehleranzahl pro Abteilung.


```python
# 1. Grundlegende statistische Analyse
basic_stats = df.describe().round(2)
basic_stats
```

### **Kategoriale Daten analysieren:**
**Geschlechterverteilung:** Prozentsatz männlich vs. weiblich. \
**Abteilungsverteilung:** Häufigkeit der verschiedenen Abteilungen. \
**Beispielanalyse:** Anzahl der Mitarbeitenden in jeder Abteilung und deren durchschnittliche Produktionszeit

```python
# 2. Geschlechterverteilung und Abteilungsverteilung
gender_distribution = df['Geschlecht'].value_counts(normalize=True) * 100
department_distribution = df['Abteilung'].value_counts()
```

```python
print(gender_distribution)
```

```python
print(department_distribution)
```

### **Datenvisualisierung:**
**Histogramme** für numerische Variablen, um die Verteilung darzustellen (z. B. Produktionszeit).


```python
# Histogramm der Produktionszeit
plt.figure(figsize=(8, 6))
sns.histplot(df['Produktionszeit'], bins=5, kde=True, color='orange')
plt.title("Histogramm der Produktionszeit")
plt.xlabel("Produktionszeit (Stunden)")
plt.ylabel("Häufigkeit")
plt.show()
```

**Balkendiagramme** für kategoriale Daten wie Geschlecht und Abteilung.

```python
# Balkendiagramm der Geschlechterverteilung
plt.figure(figsize=(8, 6))
gender_distribution.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title("Geschlechterverteilung")
plt.xlabel("Geschlecht")
plt.ylabel("Anteil (%)")
plt.show()

```

```python
"""
Männlich: 55%
Weiblich: 45%
"""
```

**Balkendiagramm der Abteilungsverteilung** zeigt die Anzahl der Mitarbeitenden in jeder Abteilung.

```python
# Balkendiagramm der Abteilungsverteilung
plt.figure(figsize=(8, 6))
department_distribution.plot(kind='bar', color='orange')
plt.title("Abteilungsverteilung")
plt.xlabel("Abteilung")
plt.ylabel("Anzahl")
plt.show()
```


**Boxplot der Produktionszeit pro Abteilung** zeigt die Verteilung und Ausreißer der Produktionszeit je Abteilung..


```python
# Boxplot der Produktionszeit pro Abteilung
plt.figure(figsize=(8, 6))
sns.boxplot(x='Abteilung', y='Produktionszeit', data=df, palette="Set2")
plt.title("Produktionszeit pro Abteilung")
plt.xlabel("Abteilung")
plt.ylabel("Produktionszeit (Stunden)")
plt.show()
```

## Tutorial: Datenanalyse mit Pandas und Matplotlib

In diesem Tutorial wirst du lernen, wie man einfache Datenauswertungen mit **Pandas** und **Matplotlib** durchführt. Wir werden Daten in einem DataFrame einlesen, grundlegende Statistiken berechnen und Grafiken erstellen.






### 1. Importieren der Bibliotheken

Wir beginnen damit, die notwendigen Bibliotheken zu importieren.

```python
# Importieren der benötigten Bibliotheken
import pandas as pd
import matplotlib.pyplot as plt
```

### 2. Erstellen eines DataFrames 
Nun erstellen wir einen einfachen DataFrame. In der Praxis würdest du oft Daten aus einer CSV-Datei oder einer Datenbank laden, aber für dieses Tutorial erstellen wir die Daten direkt.

```python
# Erstellen eines einfachen DataFrames
daten = {
    'Monat': ['Januar', 'Februar', 'März', 'April', 'Mai'],
    'Umsatz': [1500, 2200, 1800, 2400, 3000]
}

df = pd.DataFrame(daten)
df
```

### 3. Berechnen von Statistiken
Wir wollen nun einige grundlegende Statistiken über unseren Umsatz berechnen.

```python
# Berechnen des Durchschnittsumsatzes
durchschnitt = df['Umsatz'].mean()
print(f'Durchschnittlicher Umsatz: {durchschnitt} €')

# Maximaler Umsatz
max_umsatz = df['Umsatz'].max()
print(f'Maximaler Umsatz: {max_umsatz} €')
```

### 4. Plotten der Daten
Nun visualisieren wir die Daten mit einem einfachen Liniendiagramm.

```python
# Plotten eines Liniendiagramms
plt.plot(df['Monat'], df['Umsatz'], marker='o')
plt.title('Umsatz pro Monat')
plt.xlabel('Monat')
plt.ylabel('Umsatz in €')
plt.grid(True)
plt.show()
```

### 5. Hinzufügen eines Balkendiagramms
Zusätzlich zum Liniendiagramm fügen wir noch ein Balkendiagramm hinzu, um den Umsatz besser zu vergleichen.

```python
# Plotten eines Balkendiagramms
plt.bar(df['Monat'], df['Umsatz'], color='skyblue')
plt.title('Umsatz pro Monat (Balkendiagramm)')
plt.xlabel('Monat')
plt.ylabel('Umsatz in €')
plt.show()
```

### 6. Exportieren der Ergebnisse
Schließlich exportieren wir den DataFrame in eine CSV-Datei.

```python
# Exportieren des DataFrames in eine CSV-Datei
df.to_csv('umsatz.csv', index=False)
```
